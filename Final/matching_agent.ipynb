````python
from collections.abc import Mapping

def flatten_allowed_labels(keyword_categories):
    from collections.abc import Mapping

    flat_map = {}

    # Categories that should NOT be deeply traversed
    stop_deepening = {"age", "household income classification"}

    def collect_deepest_labels(node):
        labels = set()

        def recurse(subnode):
            if isinstance(subnode, Mapping):
                for k, v in subnode.items():
                    if isinstance(v, list):
                        labels.add(k.lower().replace(" ", "_"))
                    elif isinstance(v, Mapping):
                        # Keep recursing deeper
                        recurse(v)

        recurse(node)
        return labels

    def collect_top_level_labels(node):
        return set(k.lower().replace(" ", "_") for k in node)

    for cat, node in keyword_categories.items():
        key = cat.lower().replace(" ", "_")
        if cat.lower() in stop_deepening:
            labels = collect_top_level_labels(node)
        else:
            labels = collect_deepest_labels(node)
        flat_map[key] = sorted(labels)

    return flat_map
````

````python
def format_allowed_labels_text(flattened_map):
    """Format the allowed labels for the GPT prompt"""
    formatted = []
    for cat in categories:
        cat_key = cat.lower().replace(' ', '_')
        allowed = flattened_map.get(cat_key, ["N/A"])
        formatted.append(f"{cat}: {', '.join(allowed)}")
    return "\n".join(formatted)
````

````python
import ast

def ask_gpt_for_labels(question, allowed_text, flattened_map):
    prompt = (
        "You are a labeling assistant. For the given sentence, extract a label for each of the following 15 demographic categories in this exact order.\n"
        "Use ONLY the exact allowed labels provided per category. If no relevant information is mentioned, return 'N/A'.\n"
        "IMPORTANT: Use single underscores (_) to replace spaces in multi-word labels.\n"
        "IMPORTANT: Use only the main category labels provided, not subcategories or descriptive terms.\n\n"
        f"{allowed_text}\n\n"
        f"Sentence: {question}\n"
        f"Return a valid Python list with exactly 15 elements, no explanations.\n"
        f"Categories in order: {categories}\n\n"
        "Example format: ['male', 'adult', 'N/A', 'white', 'usa', 'N/A', 'N/A', 'english', 'bachelors', 'N/A', 'christian', 'married', 'engineer', 'middle_class', 'owned']\n\n"
        "Labels:"
    )

    response = client.chat.completions.create(
        model="gpt-4-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1  # Lower temperature for more consistent outputs
    )

    raw = response.choices[0].message.content.strip()
    if raw.startswith("```"):
        raw = raw.strip("`\n python")

    try:
        labels = ast.literal_eval(raw)
        if not isinstance(labels, list):
            raise ValueError("Output is not a list")
    except Exception as e:
        print(f"‚ö†Ô∏è Parsing failed: {e}\nRaw Output:\n{raw}")
        return ["N/A"] * len(categories)

    # Handle multi-language detection
    language_index = 7  # "languages spoken"
    languages_map = flattened_map.get('languages_spoken', [])
    allowed_langs = set(languages_map)
    question_lower = question.lower()

    mentioned_langs = [
        lang for lang in allowed_langs
        if lang.replace("_", " ") in question_lower or lang in question_lower
    ]

    if mentioned_langs:
        labels[language_index] = '_'.join(sorted(set(mentioned_langs)))
    else:
        labels[language_index] = 'N/A'

    # Final cleanup
    labels = clean_labels(labels)

    return labels
````

````python
def clean_labels(labels):
    """Clean and normalize labels"""
    new = []
    for i, label in enumerate(labels):
        if not isinstance(label, str):
            label = str(label)
        
        label = label.strip().lower()

        # Replace multiple spaces or underscores with a single underscore
        import re
        label = re.sub(r'[_\s]+', '_', label)
        label = label.strip('_')

        new.append(label if label else "N/A")

    # Special case: join multiple languages into one label
    language_index = 7  # 0-based index for "languages spoken"
    if isinstance(labels[language_index], list):
        cleaned_languages = [
            re.sub(r'[_\s]+', '_', str(lang).strip().lower())
            for lang in labels[language_index]
            if isinstance(lang, str) and lang.strip()
        ]
        new[language_index] = '_'.join(cleaned_languages) if cleaned_languages else 'N/A'

    return new
````

````python
def generate_labels(input_file, output_file, keyword_categories):
    df = pd.read_csv(input_file)
    if 'question' not in df.columns:
        raise ValueError("CSV must contain a 'question' column")

    flat_map = flatten_allowed_labels(keyword_categories)
    allowed_text = format_allowed_labels_text(flat_map)
    
    print("Allowed labels per category:")
    print(allowed_text)
    print("\n" + "="*50 + "\n")

    all_labels = []
    for idx, question in enumerate(df['question']):
        print(f"Processing row {idx+1}/{len(df)}...")
        labels = ask_gpt_for_labels(question, allowed_text, flat_map)
        all_labels.append(labels)
        print(f"Generated labels: {labels}")
        time.sleep(1)

    df['labels'] = all_labels
    df.to_csv(output_file, index=False)
    print(f"‚úÖ Saved labeled output to: {output_file}")


if __name__ == "__main__":
    generate_labels(
        input_file="synthetic_patient_descriptions_and_ground_truth.csv",
        output_file="labeled_output_matching_agent.csv",
        keyword_categories=keyword_categories
    )
````

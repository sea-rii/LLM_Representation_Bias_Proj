```python
import pandas as pd
import re 

def extract_age(question):
    """
    Extracts age only if it appears in an age-related phrase (e.g., '18 years old', 'aged 30').
    """
    age_patterns = [
        r'(\d{1,3})\s*years?\s*old',  # Matches "71 years old", "5 years old", etc.
        r'aged\s*(\d{1,3})',  # Matches "aged 71"
        r'(\d{1,3})-year-old',  # Matches "71-year-old"
        r'(\d{1,3})\s*yrs?\s*old',  # Matches "71 yrs old"
        r'(\d{1,3})\s*yo',  # Matches "71 yo"
        r'(\d{1,3})\s*years?',  # Matches "71 years"
        r'(\d{1,3})\s*yrs?',  # Matches "71 yrs"
    ]

    for pattern in age_patterns:
        match = re.search(pattern, question)
        if match:
            age = int(match.group(1))  # Extract number from matched pattern
            return age
    return None  # No valid age found

# Function to classify age
def classify_age(question):
    """
    Determines the correct age category if a valid age is found in the text.
    """
    age = extract_age(question)
    if age is not None:
        for category, age_range in keyword_categories['age'].items():
            if age in age_range:
                return category
    return 'N/A'  # Default if no valid age found
```

```python
# Function to categorize a question based on predefined keywords
# Function to categorize a question based on predefined keywords
def categorize_question(question, keyword_categories, label_size=15):
    """
    Assigns a single subcategory per category based on keyword matches.

    Parameters:
    - question (str): The input question text.
    - keyword_categories (dict): Dictionary containing categories and subcategories.
    - label_size (int): The fixed size of the label output.

    Returns:
    - list: A list of assigned category labels.
    """
    if pd.isna(question) or not isinstance(question, str):
        return ['N/A'] * label_size  # Handle NaN or non-string values

    question = question.lower()
    labels = []  # Initialize the labels list

    matched_categories = 0  # Initialize counter for matched categories

    for category, subcategories in keyword_categories.items():
        assigned_label = 'N/A'  # Default label

        if category == 'age':  
            assigned_label = classify_age(question)  # Handle age separately
            if assigned_label != 'N/A':
                matched_categories += 1  # Increment for age match
        elif isinstance(subcategories, dict):  # Ensure subcategories is a dictionary
            for subcategory, keywords in subcategories.items():
                if any(re.search(r'\b' + re.escape(keyword) + r'\b', question) for keyword in keywords):
                    assigned_label = subcategory
                    matched_categories += 1  # Increment the matched categories counter
                    break  # Assign first matching subcategory and exit loop

        labels.append(assigned_label)  # Append to the label list
        
    # Adjust label size dynamically
    required_size = max(len(keyword_categories), label_size)
    labels = labels[:required_size]  # Truncate if more than label_size
    labels += ['N/A'] * (required_size - len(labels))  # Fill with 'N/A' if fewer

    return labels, matched_categories

# Function to process a CSV file
def process_csv(input_file, output_file, keyword_categories):
    """
    Reads a CSV file, applies labeling to the 'question' column, and saves a new CSV.

    Parameters:
    - input_file (str): Path to the input CSV file.
    - output_file (str): Path to save the processed CSV file.
    - keyword_categories (dict): Dictionary of keyword categories.
    """
    # Load CSV file
    try:
        df = pd.read_csv(input_file)
    except FileNotFoundError:
        print(f"Error: The file {input_file} was not found.")
        return

    # Ensure the 'question' column exists
    if 'question' not in df.columns:
        print("Error: The 'question' column is missing in the input CSV.")
        return

    total_categories_matched = 0  # Initialize total matched categories counter
    total_questions = len(df)  # Total number of questions

    # Apply labeling function and count matched categories
    df['labels_and_match_count'] = df['question'].apply(
        lambda q: categorize_question(q, keyword_categories)
    )

    # Extract labels and matched categories
    df[['labels', 'matched_categories']] = pd.DataFrame(df['labels_and_match_count'].tolist(), index=df.index)

     # Remove the temporary 'labels_and_match_count' column
    df.drop(columns=['labels_and_match_count'], inplace=True)

    # Calculate total matched categories
    total_categories_matched = df['matched_categories'].sum()

    # Calculate the percentage of recognized categories
    recognition_percentage = (total_categories_matched / (total_questions * len(keyword_categories))) * 100

    # Save the results to a new CSV file
    df.to_csv(output_file, index=False)

     # Print the results
    print(f"Recognition Percentage: {recognition_percentage:.2f}%")
    print(df[['question', 'labels', 'matched_categories']].head())  # Display the first 5 rows as a preview

# Example usage
input_csv = 'medhalt_causal_new.csv'
output_csv = 'new_output_labels.csv'
process_csv(input_csv, output_csv, keyword_categories)
```
